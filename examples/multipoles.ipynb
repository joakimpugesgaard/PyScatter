## Description
This class implements the multipolar decomposition in the formalism outlined by M.E. Rose. By construction, the multipolar fields are defined in the parity basis (magnetic and electric types), having left circular, right circular or z polarizations. 

## Initialization

Here, it is demonstrated how to instantiate a multipoles object, using a predefined domain object, along with the wavelength and the multipole order. Furthermore, the get_multipoles() method is shown, which is useful for custom expansions

```python
from src.multipoles import *

# initialize domain
size = 1.5
planes = {'xy', 'xz'}
points = 150
d = domain(size, points, planes)

# initialize multipole object
mp = multipoles(l_max = 5, m_max = 5, wavelength = 0.632, domain = d)

# calculate a multipole with l=m=3 
mpp = mp.get_multipoles(l = 3, m = 3, spatial_fun = "bessel")

# display multipole structure
print(mpp.keys())
```

## Plots

The plot_multipoles() method can be used to plot a multipole of certain order, using the l and m arguments. Furthermore the type is specified with a string, being either "electric" or "magnetic".    
For an outgoing type, use the "scattering" interaction (this will exclude a central region from the plot, while using a Hankel function for the outer region). Otherwise, use "internal".   
Plots can be separated in polatization components or only in planes. Finally, with the "globalnorm" argument, plots can be forced onto a global scale to see the magnitudees more clearly.

```python
# plot multipole in polarization components, using the scattering type (outgoing / hankel function)
mp.plot_multipoles(l = 2, m = 2, type = "electric", interaction = "scattering", plot = "components", globalnorm = True)
```

```python
# plot the total field of magnetic multipole using the internal type (standing/ bessel)
mp.plot_multipoles(l = 2, m = 0, type = "magnetic", interaction = "internal", plot = "total", globalnorm = True)
```

```python
# plot total intensity both using both the scattering and internal type, 
# using a global normalization for proper scaling
mp.plot_multipoles(l = 2, m = 0, type = "magnetic", interaction = "both", plot = "total", globalnorm = True)
```
