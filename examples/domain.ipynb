## Description
The domain class is a foundational class used by all other classes to perform spatial evaluations on (in most cases) spherical coordinate system. The domains are, however, defined in cartesian coordinates to be more intuitive. This notebook demonstrates how to set up a domain object and how to use its methods.

## Initialization    
Below is how to define a domain object, which is often a necessary input to create more advanced objects. 

```python
from src.domain import domain

#set domain size (width of domain)

size = 2 #Âµm
points = 150 # resolution

#planes of interest
planes = {'xy', 'yz'}

#initiate domain object
d = domain(size, points, planes)
```

## Visulization   
While the methods of this class are not very enlightening on their own, it is useful to check the definitions of the spherical coordinates by displaying them. this is shown below.

```python
#transform grids in spherical coordinates
spherical_grids = d.spherical_grid()

coords = ['r', 'theta', 'phi']

#show coordinates for sanity check
for coord in coords:
    print(f'Showing {coord} grid:')
    d.show_coord(coord, spherical_grids)
```
