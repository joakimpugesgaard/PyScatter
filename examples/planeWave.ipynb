## Description

The planeWave module implements the multipolar expansion of a plane wave in the parity basis, allowing it to be applied with the 'interaction' module for scattering problems. Since plane waves are a standard assumption in many theoretical studies, it is a useful tool to use along the focused beams for comparisons.    


## Initialization

For initializing the planeWave object, a domain object again needs to be defined. The planeWave object furthermore needs a wavelength and a polarization, which can be either $x$, $y$, left or right circular.

```python
from src.planeWave import *
```

```python
# initialize domain
size = 3
planes = {'xy', 'yz'}
d = domain(size, planes=planes)
wavelength = 1 # Âµm
```

```python
# initialize planewave object left right circular polarization

pw = planeWave(wavelength=wavelength, domain=d, polarization="left circular")
```

Using the 'plotBeam()' method, the norm of the real part of the field is shown - either in polarization components (plot = "components") or in total (plot = "total"). This is simple to see the wavelength relative to the domain size.

```python
# display beam in polarization components
pw.plotBeam(plot = "components", globalnorm = False)   
```

Notice that all slices in the propagation direction (z) are constant, as they should be. Moreover, the wave is clearly only left circular polarized. Defining an x-polarized PW yields the following result, which looks the same, since the basis in the plot is changed. 
If one were to plot this PW in the circular polarization basis, the field would be split between the left and right circular polarization.

```python
pw_lin = planeWave(wavelength=wavelength, domain=d, polarization="x")
pw_lin.plotBeam(plot = "components", globalnorm = False)   
```

```python
# plot total field. 
pw.plotBeam(plot = "total", globalnorm = True)
```
